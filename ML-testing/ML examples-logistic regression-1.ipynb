```python
from sklearn.datasets import fetch_openml
mnist=fetch_openml('mnist_784')

```

```python
#Pickling the mnist data once downloaded from server

import pickle
with open('data.pkl','wb') as pickle_file:
    pickle.dump(mnist,pickle_file)
    
```

```python
# Retriving the pickle file
with open('data.pkl','rb') as pickle_file:
    mnist_new=pickle.load(pickle_file)


x,y=mnist_new['data'],mnist_new['target']
X_train,X_test=x[0:60000],x[60000:70000]
Y_train,Y_test=y[0:60000],y[60000:70000]







```

```python
import matplotlib
import matplotlib.pyplot as plt
some_image=x[1201]
image_sample=some_image.reshape(28,28)
plt.imshow(image_sample,cmap=matplotlib.cm.binary)
# y[1201]
```

```python
import numpy as np
suffle_index=np.random.permutation(60000)
X_train,Y_train=X_train[suffle_index],Y_train[suffle_index]
Y_train=Y_train.astype(np.int8)
Y_test=Y_test.astype(np.int8)
```

## Here im creating a classifier using logistic regression to find the given test data is 6 or not


```python

Y_train2=(Y_train==6)
Y_test2=(Y_test==6)
Y_train2.shape       



```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(X_train,Y_train2)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(clf,X_test,Y_test2,cv=2,scoring="accuracy")
```
