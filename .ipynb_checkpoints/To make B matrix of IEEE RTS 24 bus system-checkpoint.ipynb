```python
#Not to run this file again unless the Bmatrix formed is needed to be re evaluated,otherwise the program will overwrite the
#Bmatix csv file formed



import pandas as pd
import numpy as np


#loading the excel file
data1=pd.read_csv("f3.csv")


rows, cols = (25,25) 
  
B =[[0 for x in range(rows)] for y in range(cols)] 



fbus=data1['fbus']  
x_reci=data1['1/x']
tbus=data1['tbus']


for i in range (0,38):
        B[fbus[i]-1][tbus[i]-1]=-x_reci[i]
        B[tbus[i]-1][fbus[i]-1]=-x_reci[i]
        
    
        
print(B)        

#Converting the matrix into pandas DF and writing it to the CSV file

pd.DataFrame(B).to_csv('Bmatrix.csv')

```
